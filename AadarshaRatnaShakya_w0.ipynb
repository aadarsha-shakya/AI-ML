```python
def length_conversion(value: float, unit: str) -> float:
    if unit == 'm':
        return value * 3.28084  # meters to feet
    elif unit == 'ft':
        return value / 3.28084  # feet to meters
    else:
        raise ValueError('Unsupported length unit.')

def weight_conversion(value: float, unit: str) -> float:
    if unit == 'kg':
        return value * 2.20462  # kilograms to pounds
    elif unit == 'lbs':
        return value / 2.20462  # pounds to kilograms
    else:
        raise ValueError('Unsupported weight unit.')

def volume_conversion(value: float, unit: str) -> float:
    if unit == 'L':
        return value * 0.264172  # liters to gallons
    elif unit == 'gal':
        return value / 0.264172  # gallons to liters
    else:
        raise ValueError('Unsupported volume unit.')


def main():
    try:
        conversion_type = input('Choose conversion type (length, weight, volume): ').strip().lower()
        value = float(input('Enter the value to convert: '))

        if conversion_type == 'length':
            unit = input('Enter the unit (m or ft): ').strip().lower()
            result = length_conversion(value, unit)
            print(f'Converted value: {result}')
        elif conversion_type == 'weight':
            unit = input('Enter the unit (kg or lbs): ').strip().lower()
            result = weight_conversion(value, unit)
            print(f'Converted value: {result}')
        elif conversion_type == 'volume':
            unit = input('Enter the unit (L or gal): ').strip().lower()
            result = volume_conversion(value, unit)
            print(f'Converted value: {result}')
        else:
            print('Invalid conversion type.')
    except ValueError as e:
        print(f'Error: {e}')

if __name__ == '__main__':
    main()
```

```python
def calculate_sum(numbers: list) -> float:
    return sum(numbers)

def calculate_average(numbers: list) -> float:
    if not numbers:
        raise ValueError('Cannot calculate average of an empty list.')
    return sum(numbers) / len(numbers)

def calculate_maximum(numbers: list) -> float:
    return max(numbers)

def calculate_minimum(numbers: list) -> float:
    return min(numbers)


def main():
    try:
        operation = input('Choose operation (sum, average, maximum, minimum): ').strip().lower()
        numbers_input = input('Enter a list of numbers separated by spaces: ').strip()
        numbers = [float(num) for num in numbers_input.split()]
        if not numbers:
            raise ValueError('The list of numbers cannot be empty.')

        if operation == 'sum':
            result = calculate_sum(numbers)
        elif operation == 'average':
            result = calculate_average(numbers)
        elif operation == 'maximum':
            result = calculate_maximum(numbers)
        elif operation == 'minimum':
            result = calculate_minimum(numbers)
        else:
            print('Invalid operation. Please choose from sum, average, maximum, minimum.')
            return
        print(f'Result: {result}')
    except ValueError as e:
        print(f'Error: {e}')

if __name__ == '__main__':
    main()
```

```python
#Extract Every Other Element:
def extract_every_other(lst: list) -> list:
    return lst[::2]
#Slice a Sublist:
def get_sublist(lst: list, start: int, end: int) -> list:
    return lst[start:end+1]
#Reverse a List Using Slicing:
def reverse_list(lst: list) -> list:
    return lst[::-1]
#Remove the First and Last Elements:
def remove_first_last(lst: list) -> list:
    return lst[1:-1]
#Get the First n Elements:
def get_first_n(lst: list, n: int) -> list:
    return lst[:n]
#Extract Elements from the End:
def get_last_n(lst: list, n: int) -> list:
    return lst[-n:]
#Extract Elements in Reverse Order:
def reverse_skip(lst: list) -> list:
    return lst[-2::-2]

# Example usage
if __name__ == '__main__':
    lst = [1, 2, 3, 4, 5, 6]
    print("Extract Every Other Element:", extract_every_other(lst))
    print("Slice a Sublist (2 to 4):", get_sublist(lst, 2, 4))
    print("Reverse a List:", reverse_list(lst))
    print("Remove First and Last Elements:", remove_first_last(lst))
    print("Get First 3 Elements:", get_first_n(lst, 3))
    print("Get Last 2 Elements:", get_last_n(lst, 2))
    print("Extract Elements in Reverse Order:", reverse_skip(lst))
```

```python
from typing import Any, List

def flatten(lst: List) -> List:
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list

def access_nested_element(lst: List, indices: List[int]) -> Any:
    element = lst
    for index in indices:
        element = element[index]
    return element

def sum_nested(lst: List) -> float:
    return sum(flatten(lst))

def remove_element(lst: List, elem: Any) -> List:
    if isinstance(lst, list):
        return [remove_element(sub, elem) for sub in lst if sub != elem]
    return lst

def find_max(lst: List) -> Any:
    return max(flatten(lst))

def count_occurrences(lst: List, elem: Any) -> int:
    return flatten(lst).count(elem)

def deep_flatten(lst: List) -> List:
    return flatten(lst)

def average_nested(lst: List) -> float:
    flat_list = flatten(lst)
    return sum(flat_list) / len(flat_list) if flat_list else 0

# Example usage
if __name__ == '__main__':
    nested_list = [[1, 2], [3, [4, 5]], 6]
    print("Flatten a Nested List:", flatten(nested_list))
    print("Access Nested Element:", access_nested_element([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [1, 2]))
    print("Sum of All Elements:", sum_nested(nested_list))
    print("Remove Specific Element:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
    print("Find Maximum Element:", find_max(nested_list))
    print("Count Occurrences:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
    print("Deep Flatten a List of Lists of Lists:", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
    print("Nested List Average:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

```python
import numpy as np

# Given arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^(-1) = I (Identity Matrix)
A_inv = np.linalg.inv(A)  # Inverse of A
I = np.eye(2)  # Identity matrix of size 2x2
result_1 = np.dot(A, A_inv)
print("A * A^(-1) =\n", result_1)
print("Identity Matrix:\n", I)

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB =\n", AB)
print("BA =\n", BA)

# Check if AB == BA
if np.array_equal(AB, BA):
    print("\nAB is equal to BA")
else:
    print("\nAB is not equal to BA")

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("\n(AB)^T =\n", AB_T)
print("B^T * A^T =\n", BT_AT)

# Check if (AB)^T == B^T * A^T
if np.array_equal(AB_T, BT_AT):
    print("\n(AB)^T = B^T * A^T is True")
else:
    print("\n(AB)^T = B^T * A^T is False")

# Solve the linear system using Inverse Method

# Representing the equations in matrix form as AX = B
# The system of equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Coefficient matrix A
A_system = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])

# Constant matrix B
B_system = np.array([-1, -3, 9])

# Solving for X using inverse of A
X = np.linalg.inv(A_system).dot(B_system)
print("\nSolution to the linear system (Inverse Method):\n", X)

# Now solve using np.linalg.solve (Alternative to using np.linalg.inv)
X_solved = np.linalg.solve(A_system, B_system)
print("\nSolution using np.linalg.solve:\n", X_solved)
```
